```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r'C:\Users\jasmi\Projects\Project_1\Project-1-Elysium\Tulsa_AQI_data.csv'
tulsa_aqi_data = pd.read_csv(file_path)

# Use 'arithmetic_mean' as the AQI measurement
tulsa_aqi_filtered = tulsa_aqi_data[['Year', 'quarter', 'arithmetic_mean']]

# Create a complete DataFrame with all combinations of years and quarters
years = range(2014, 2024)
quarters = [1, 2, 3, 4]
complete_index = pd.MultiIndex.from_product([years, quarters], names=['Year', 'quarter'])
complete_df = pd.DataFrame(index=complete_index).reset_index()

# Group by Year and Quarter, then calculate the mean AQI for each group
grouped_data = tulsa_aqi_filtered.groupby(['Year', 'quarter']).mean().reset_index()

# Merge the complete DataFrame with the grouped data to ensure all combinations are present
merged_data = pd.merge(complete_df, grouped_data, on=['Year', 'quarter'], how='left')

# Fill missing values using interpolation
merged_data['arithmetic_mean'] = merged_data['arithmetic_mean'].interpolate()

# Create a new column for plotting that combines year and quarter
merged_data['Year_Quarter'] = merged_data['Year'] + (merged_data['quarter'] - 1) / 4

# Plotting the averaged data
plt.figure(figsize=(12, 6))

# Plot a continuous line for all averaged data
plt.plot(merged_data['Year_Quarter'], merged_data['arithmetic_mean'], linestyle='-', color='black', label='AQI')

# Overlay markers for each quarter
markers = ['o', 'o', 'o', 'o']  # Different markers for each quarter
colors = ['blue', 'orange', 'green', 'red']

for i, quarter in enumerate(quarters):
    quarter_data = merged_data[merged_data['quarter'] == quarter]
    plt.scatter(quarter_data['Year_Quarter'], quarter_data['arithmetic_mean'], 
                color=colors[i], marker=markers[i], label=f'Quarter {quarter}')

# Set the x-axis ticks to include every year
plt.xticks(ticks=range(2014, 2024), labels=[str(year) for year in range(2014, 2024)])

plt.title('Quarterly Average Arithmetic Mean AQI in Tulsa (2014-2023)')
plt.xlabel('Year')
plt.ylabel('Average Arithmetic Mean AQI')
plt.legend(title='Quarter')
plt.grid(True)
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
tulsa_aqi_data = pd.read_csv(r'C:\Users\jasmi\Projects\Project_1\Project-1-Elysium\Tulsa_AQI_data.csv')
la_aqi_data = pd.read_csv(r'C:\Users\jasmi\Projects\Project_1\Project-1-Elysium\LA_AQI_data.csv')
ny_aqi_data = pd.read_csv(r'C:\Users\jasmi\Projects\Project_1\Project-1-Elysium\NY_AQI_data.csv')

def prepare_data(aqi_data):
    # Use 'arithmetic_mean' as the AQI measurement
    aqi_filtered = aqi_data[['Year', 'quarter', 'arithmetic_mean']]
    
    # Create a complete DataFrame with all combinations of years and quarters
    years = range(2014, 2024)
    quarters = [1, 2, 3, 4]
    complete_index = pd.MultiIndex.from_product([years, quarters], names=['Year', 'quarter'])
    complete_df = pd.DataFrame(index=complete_index).reset_index()
    
    # Group by Year and Quarter, then calculate the mean AQI for each group
    grouped_data = aqi_filtered.groupby(['Year', 'quarter']).mean().reset_index()
    
    # Merge the complete DataFrame with the grouped data to ensure all combinations are present
    merged_data = pd.merge(complete_df, grouped_data, on=['Year', 'quarter'], how='left')
    
    # Fill missing values using interpolation
    merged_data['arithmetic_mean'] = merged_data['arithmetic_mean'].interpolate()
    
    # Create a new column for plotting that combines year and quarter
    merged_data['Year_Quarter'] = merged_data['Year'] + (merged_data['quarter'] - 1) / 4
    
    return merged_data

# Prepare the data for each city
tulsa_data = prepare_data(tulsa_aqi_data)
la_data = prepare_data(la_aqi_data)
ny_data = prepare_data(ny_aqi_data)

# Plotting the data for all three cities with quarterly markers on the same line
plt.figure(figsize=(12, 6))

# Define markers for each quarter
markers = ['o', 'o', 'o', 'o']  # Different markers for each quarter

# Plot Tulsa data with markers for quarters
plt.plot(tulsa_data['Year_Quarter'], tulsa_data['arithmetic_mean'], linestyle='-', color='blue', label='Tulsa')
for i, quarter in enumerate([1, 2, 3, 4]):
    quarter_data = tulsa_data[tulsa_data['quarter'] == quarter]
    plt.scatter(quarter_data['Year_Quarter'], quarter_data['arithmetic_mean'], 
                color='blue', marker=markers[i], label=f'Tulsa Q{quarter}' if i == 0 else "")

# Plot LA data with markers for quarters
plt.plot(la_data['Year_Quarter'], la_data['arithmetic_mean'], linestyle='-', color='orange', label='LA')
for i, quarter in enumerate([1, 2, 3, 4]):
    quarter_data = la_data[la_data['quarter'] == quarter]
    plt.scatter(quarter_data['Year_Quarter'], quarter_data['arithmetic_mean'], 
                color='orange', marker=markers[i], label=f'LA Q{quarter}' if i == 0 else "")

# Plot NY data with markers for quarters
plt.plot(ny_data['Year_Quarter'], ny_data['arithmetic_mean'], linestyle='-', color='green', label='NY')
for i, quarter in enumerate([1, 2, 3, 4]):
    quarter_data = ny_data[ny_data['quarter'] == quarter]
    plt.scatter(quarter_data['Year_Quarter'], quarter_data['arithmetic_mean'], 
                color='green', marker=markers[i], label=f'NY Q{quarter}' if i == 0 else "")

# Set the x-axis ticks to include every year
plt.xticks(ticks=range(2014, 2024), labels=[str(year) for year in range(2014, 2024)])

plt.title('Quarterly Average Arithmetic Mean AQI (2014-2023)')
plt.xlabel('Year')
plt.ylabel('Average Arithmetic Mean AQI')
plt.legend(title='City and Quarter')
plt.grid(True)
plt.show()

```
