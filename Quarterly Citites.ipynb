```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
tulsa_aqi_data = pd.read_csv(r'C:\Users\jasmi\Projects\Project_1\Project-1-Elysium\Tulsa_AQI_data.csv')
la_aqi_data = pd.read_csv(r'C:\Users\jasmi\Projects\Project_1\Project-1-Elysium\LA_AQI_data.csv')
ny_aqi_data = pd.read_csv(r'C:\Users\jasmi\Projects\Project_1\Project-1-Elysium\NY_AQI_data.csv')

def prepare_data(aqi_data):
    # Use 'arithmetic_mean' as measurement
    aqi_filtered = aqi_data[['Year', 'quarter', 'arithmetic_mean']]
    
    # DataFrame with all years and quarters
    years = range(2014, 2024)
    quarters = [1, 2, 3, 4]
    complete_index = pd.MultiIndex.from_product([years, quarters], names=['Year', 'quarter'])
    complete_df = pd.DataFrame(index=complete_index).reset_index()
    
    # Group by Year and Quarter, calculate the mean
    grouped_data = aqi_filtered.groupby(['Year', 'quarter']).mean().reset_index()
    
    # Merge the DataFrame
    merged_data = pd.merge(complete_df, grouped_data, on=['Year', 'quarter'], how='left')
    
    # Create a new column
    merged_data['Year_Quarter'] = merged_data['Year'] + (merged_data['quarter'] - 1) / 4
    
    return merged_data

# Prepare the data for each city
tulsa_data = prepare_data(tulsa_aqi_data)
la_data = prepare_data(la_aqi_data)
ny_data = prepare_data(ny_aqi_data)

# tulsa_data.head()
# Plotting the data for all three cities with quarterly markers on the same line
plt.figure(figsize=(12, 6))

# Plot Tulsa
plt.plot(tulsa_data['Year_Quarter'], tulsa_data['arithmetic_mean'], linestyle='-', color='red', label='Tulsa')
    

# Plot LA
plt.plot(la_data['Year_Quarter'], la_data['arithmetic_mean'], linestyle='-', color='blue', label='LA')
   

# Plot NY
plt.plot(ny_data['Year_Quarter'], ny_data['arithmetic_mean'], linestyle='-', color='green', label='NY')
   
# Plot
plt.xticks(ticks=range(2014, 2024), labels=[str(year) for year in range(2014, 2024)])
plt.title('Quarterly Average Arithmetic Mean AQI (2014-2023)')
plt.xlabel('Year')
plt.ylabel('Average Arithmetic Mean AQI')
plt.legend(title='City and Quarter')
plt.grid(True)
plt.show()

```
